Bu ödevimizde bizden istenen hedef, Fashion_MNIST datasındaki pixel değerlerini tek boyuttan iki boyutlu düzleme taşıyıp bir matris oluşturmak ve bu matriste yazan değerleri görselleştirmek. Elimizde n satırlı(69999), 785 sütunlu bir veri var ve her bir satır bir kıyafet görselinin pixellerini barındırıyor. Her bir satırda 785 adet kolon mevcut ve son kolona kadar pixel numaraları kolonları oluşturuyor. Yani aslında her bir görsel için 784 adet pixelimiz mevcut. Bunu 2 boyuta yansıttığımızda elimize 28x28'lik bir matris geçiyor ve bu şekilde görselimizi görüntüleyebiliyoruz. Ayrıca ödevde bizden 32x32'lik bir matris daha isteniyor ve bunu yapmak için kullandığım yöntem Padding oldu. Padding'e sınır ekleme mantığı ile yaklaştım(daha önceden Frontend tasarım yaparken sıkça karşılaştığım bir terim olduğundan dolayı bende böyle bir çağrışım uyandırdı) ve bu şekilde ana matrisimin sınırlarına yeni matrisimde oluşturmak istediğim satır sayısında sınır eklersem hedefime ulaşacaktım. 28x28 matrisimi 32x32 yapmak için esas verimin içerisinde barınan her bir görseli oluşturan arraylerin çevresini ((2,2),(2,2)) saracak şekilde yeni değerler ekledim ve sonuç olarak 32x32 bir matris elde etmiş oldum. Bu anlatımın basitçe örneklendirilmiş hali aşağıdaki görselde bulunmakta.

```python
def summarize_classification(y_test,
                             y_pred,
                             avg_method='weighted'):
  acc = accuracy_score(y_test, y_pred, normalize=True)
  num_acc = accuracy_score(y_test, y_pred, normalize=False)

  prec = precision_score(y_test, y_pred, average='weighted')
  recall = recall_score(y_test, y_pred, average='weighted')

  print("Test data count: ", len(y_test))
  print("accuracy_count: ", num_acc)
  print("Accuracy_score: ", acc)
  print("precision_score: ", prec)
  print("recall_score: ", recall)
```
